## Penjelasan Variable Dataset
`Status` = Developing or Developed \
`Life expectancy` = Ekspektasi umur \
`Adult Mortality` = Tingkat Kematian Orang Dewasa dari kedua jenis kelamin (probabilitas kematian antara 15 dan 60 tahun per 1000 populasi) \
`Infant Deaths` = Jumlah Kematian Bayi per 1000 penduduk \
`Alcohol` = Alkohol, tercatat konsumsi per kapita (15+) (dalam liter alkohol murni) \
`Percentage Expenditure` = Pengeluaran untuk kesehatan sebagai persentase dari Produk Domestik Bruto per kapita(%) \
`Hepatitis B` = Cakupan imunisasi Hepatitis B (HepB) pada anak usia 1 tahun (%) \
`Measles` = jumlah kasus campak yang dilaporkan per 1000 penduduk \
`BMI` = Rata-rata BMI dari seluruh populasi \
`Under five deaths` = Jumlah kematian balita per 1000 penduduk \
`Polio` = Cakupan imunisasi Polio (Pol3) pada anak usia 1 tahun (%) \
`Total expenditure` = Pengeluaran pemerintah untuk kesehatan sebagai persentase dari total pengeluaran pemerintah (%) \
`Diphteria` = Cakupan imunisasi difteri tetanus toksoid dan pertusis (DTP3) pada anak usia 1 tahun (%) \
`HIV/AIDS` = Kematian per 1.000 kelahiran HIV/AIDS (0-4 tahun) \
`GDP` = GDP pada us dollar \
`Population` = Jumlah populasi daerah \
`thinness  1-19 years` = Prevalensi kurus di kalangan anak-anak dan remaja untuk Usia 10 sampai 19 (%) \
`thinness 5-9 years` = Prevalensi kurus di kalangan anak-anak dan remaja untuk Usia 5 sampai 9 (%) \
`Income composition of resources` = Indeks Pembangunan Manusia dalam hal komposisi sumber daya pendapatan (indeks mulai dari 0 hingga 1) \
`Schooling` = Lama tahun sekolah (tahun) \
**Kemudian untuk `Outcome` adalah `Life Expectancy`**

### Mengecek Nilai Outlier

```python
# Melihat distribusi boxplot setiap features numerik.
columns_ = ["Life_Expectancy", "Adult_Mortality", "Alcohol", "Hepatitis_B", "BMI", "Polio","Total_Expenditure","Diphtheria","GDP","Population","Thinness_1-19_Years","Thinness_5-9_Years","Income_Composition_of_Resources","Schooling"]
index_ = 0
fig, ax = plt.subplots(5,3, figsize = (21,25)) # membuat figur dan axis untuk subplots
for row in range(5):
    for col in range(3):
        ax[row,col].boxplot(life[columns_[index_]])
        ax[row,col].set_title(f"Distribusi {columns_[index_]}")
        index_ += 1
plt.show()
```

```python
# Imputasi Nilai Outlier dengan nilai median
# Melihat distribusi setiap variabel numerik setelah imputasi outlier.
columns_ = ["Life_Expectancy", "Adult_Mortality", "Alcohol", "Hepatitis_B", "BMI", "Polio","Total_Expenditure","Diphtheria","GDP","Population","Thinness_1-19_Years","Thinness_5-9_Years","Income_Composition_of_Resources","Schooling"]
index_ = 0
fig, ax = plt.subplots(5,3, figsize = (21,25)) # membuat figur dan axis untuk subplots
for row in range(5):
    for col in range(3):
        # batas atas data
        q1 = life[columns_[index_]].quantile(q=0.25) 
        q3 = life[columns_[index_]].quantile(q=0.75) 
        iqr = q3-q1
        batas_atas = q3 + (iqr*1.5)
        batas_bawah = q1 - (iqr*1.5)
        # Mencari nilai median
        med = life[columns_[index_]].median()
        # mengubah nilai outlier dengan median
        life.loc[life[columns_[index_]] > batas_atas, columns_[index_]]=med 
        life.loc[life[columns_[index_]] < batas_bawah, columns_[index_]]=med
        # menampilkan viz pada fig dan ax
        ax[row,col].boxplot(life[columns_[index_]])
        ax[row,col].set_title(f"Distribusi 2 {columns_[index_]}")
        index_ += 1
plt.show()
```

Untuk outlier tidak dilakukan penghapusan melainkan mengganti nilai outlier dengan niai median. Sehingga tidak mengubah struktur data. Kemudian untuk batas bawah outlier adalah Quartil 1 - (1.5 * IQR). Batas atas adalah Quartil 3 + (1.5 * IQR). Jika data diluar batas atas dan batas bawah maka nilainya akan diganti dengan median.

## Performing Statistical Test

```python
from scipy.stats import ttest_ind,levene,kruskal
import statsmodels.api as sm
from statsmodels.formula.api import ols
from statsmodels.stats.multicomp import pairwise_tukeyhsd
from statsmodels.stats.diagnostic import normal_ad
```

```python
# Uji t test 2 sample pada variabel Status
Developed = life[life["Status"] == "Developed"]["Life_Expectancy"] # membuat var developed
Developing =  life[life["Status"] == "Developing"]["Life_Expectancy"] # membuat var developing

n_developed = Developed.shape[0] # Menghitung jumlah observasi variabel Status pada kategori developed.
n_developing = Developing.shape[0] # Menghitung jumlah observasi variabel Status pada kategori developing.

result = ttest_ind(Developed, Developing, equal_var = True, random_state = 42,alternative= 'greater') # melakukan Uji t.
```

```python
# Uji t test 2 sample pada variabel Status
print("Banyak sampel status developing = ",n_developing)
print("Banyak sampel status developed = ",n_developed)

print("-"*50)
print('Pertama, menyatakan hipotesis dan taraf signifikansi:')
print('H0 = miu_developed - miu_developing <= 0')
print('H1 = miu_developed - miu_developing > 0')
print("alpha = 0.05")

print("-"*50)
print("Statistik Hitung = ",result[0])
print("P Value = ",result[1])

print("-"*50)
print("Karna pvalue lebih kecil dari alpha, maka dapat diartikan bahwa terdapat perbedaan signifikan Life Expectancy negara status Developed dan Developing. Sehingga dapat dismpulkan menolak H0 dimana rata-rata Status negara Developed memiliki probability Life Expectancy lebih besar dari pada Status negara Developing.")
```

```python
# Uji anova pada variabel Country
print("Pertama, menyatakan hipotesis dan taraf signifikansi")
print("H0 = Semua country memiliki miu yang sama")
print("H1 = Setidaknya terdapat satu perbedaan miu")

print("-"*50)
model_country = ols('Life_Expectancy ~ Country', data =life).fit() # Membuat persamaan linear antara Life Expectancy dengan variabel Country
anova_country = sm.stats.anova_lm(model_country) # Melakuakn pengujian anova

print("alpha = 0,05")
print("Hasil pengujian test Anova")
print(anova_country)

print("-"*50)
print("Karena hasil uji ANOVA menyatakan p-value  < 0.05, maka keputusan yang diperoleh adalah menolak H0 yang dimana setidaknya ada perbedaan rata - rata probabilitas Angka harapan hidup yang signifkan pada kategori Country")
```

#### Scale Income_Composition_of_Resources 

```python
# Mengalikan variabel prediktor Income_Composition_of_Resources dengan 100
encoded['Income_Composition_of_Resources_2'] = encoded['Income_Composition_of_Resources']*100
```

```python
# Construct model 
# Create OLS model object
model1 = smf.ols("Life_Expectancy ~ Alcohol + Polio + Income_Composition_of_Resources_2 + Schooling + Status_Developed + Status_Developing", data=encoded)

# Fit the model
results2 = model1.fit()

#Show result
print(results2.summary())
```

Setelah dilakukan pemodelan didapatkan nilai koefisien seperti diatas. Proses selanjutnya adalah melakukan centering kepada variabel Alcohol, Polio, Income_Composition_of_Resources_2, dan Schooling agar saat diinterpretasikan lebih masuk akal.

### Center Variable Numerik

```python
encoded['C_Alcohol'] = encoded['Alcohol'] - np.mean(encoded['Alcohol'])
encoded['C_Polio'] = encoded['Polio'] - np.mean(encoded['Polio'])
encoded['C_Income_Composition_of_Resources_2'] = encoded['Income_Composition_of_Resources_2'] - np.mean(encoded['Income_Composition_of_Resources_2'])
encoded['C_Schooling'] = encoded['Schooling'] - np.mean(encoded['Schooling'])
encoded
```

Selanjutnya membuat model menggunakan variabel baru yang telah di centering.

```python
# Construct model 
# Create OLS model object
model2 = smf.ols("Life_Expectancy ~ C_Alcohol + C_Polio + C_Income_Composition_of_Resources_2 + C_Schooling + Status_Developed + Status_Developing", data=encoded)

# Fit the model
results2 = model2.fit()

#Show result
print(results2.summary())
```
